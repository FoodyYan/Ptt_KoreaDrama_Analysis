```python
import jieba
voice=[]
date1 = datetime.date(2021, 1, 1)
date2 = datetime.date(2021, 6, 20)
begin_date = "01/01"
old_date_temp = datetime.datetime.strptime(begin_date, "%m/%d")
current_date = "01/01"
drama_voice=0
for name in range(len(drama_list)):
    # print(name)
    for post_date in range((date2 - date1).days +1):#天數加一
        for index in range (len(df)):
            date_ = "0" + df.date[index][1:] 
            # print("im date" + date_)
            # print("Im Current_Date" + current_date)
            if  date_ == current_date:
                #print("asdfdafsdfsafafadfs")
                #計算被提及幾次
                #切詞
                words = list(jieba.cut(df.title_content[index]))
                name_index=len(drama_list[name])-1
                drama_name=drama_list[name][0]
                # print(drama_name)
                drama_voice=words.count(drama_list[name][name_index])+drama_voice
                # print(drama_name+str(drama_voice))
                if drama_voice > 0:
                    print(drama_name + "=>voice="+str(drama_voice) + df.title_content[index] + "\n")
        voice.append([drama_name,current_date,drama_voice])       
        current_date = old_date_temp + datetime.timedelta(days=post_date)
        current_date = current_date.strftime("%m/%d")
        # print("Im Current_Date"current_date)
        drama_voice=0
        #OKAY
```
